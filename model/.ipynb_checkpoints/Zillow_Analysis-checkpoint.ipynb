# Example 1: XGBoost with Supplemental Data
How do fluctuations in the S&P 500 affect median sales prices of homes in the US?

## Add SPX Information

```python
SPX = load_dataset("misikoff/SPX", download_mode="force_redownload")
```

```python
spx_df = SPX['train'].to_pandas()
spx_df['Date'] = spx_df['Date'].astype(str)

date_spx_pairings = {}
dates = input_df['Date'].unique()
starting_index = 0
for d in dates:
  str_date = str(d)[0:10]
  adj_close = float("NaN")
  for index, row in spx_df[starting_index:].iterrows():
    starting_index = index
    dif = d - np.datetime64(row['Date'])
    if dif.days < 0:
      break
    else:
      adj_close = row['Adj Close']

  date_spx_pairings[str_date] = adj_close

# date_spx_pairings
```

```python
input_df['SPX'] = input_df['Date'].astype(str).map(date_spx_pairings)
input_df.to_csv('input_df.csv', index=False)
input_df
```

## Prepare Dateframe for modeling

```python
include_SPX = True

exp_cols = [
  'Region',
  # 'State',
  'time',
  'Home Type',
  # 'Median Sale to List Ratio',
  'Month',
  # 'Date'
]
supplementary_cols = [
  'SPX'
]
target = 'Median Sale Price'

cols = exp_cols
if include_SPX:
  cols = cols + supplementary_cols

# Look at city data only
cur_df = input_df[input_df['Region Type'] == 3]
cur_df = cur_df[cols + [target]]

cur_df = cur_df.dropna()
cur_df
```

## Run Model

```python
import numpy as np

X = cur_df.drop('Median Sale Price', axis=1)
y = cur_df[['Median Sale Price']]

# cutoff = cur_df['time'].quantile(q=80/100)

# X_train = cur_df[cur_df['time'] < cutoff].drop('Median Sale Price', axis=1)
# y_train = cur_df[cur_df['time'] < cutoff][['Median Sale Price']]

# X_test = cur_df[cur_df['time'] >= cutoff].drop('Median Sale Price', axis=1)
# y_test = cur_df[cur_df['time'] >= cutoff][['Median Sale Price']]
```

```python
# cats = X_train.select_dtypes(exclude=np.number).columns.tolist()

# for col in cats:
#    X_train[col] = X_train[col].astype('category')
#    X_test[col] = X_test[col].astype('category')

cats = X.select_dtypes(exclude=np.number).columns.tolist()

for col in cats:
   X[col] = X[col].astype('category')
```

```python
X.dtypes
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

X_train
```

```python
import xgboost as xgb

# Create regression matrices
dtrain_reg = xgb.DMatrix(X_train, y_train, enable_categorical=True)
dtest_reg = xgb.DMatrix(X_test, y_test, enable_categorical=True)
```

```python
params = {"objective": "reg:squarederror", "tree_method": "auto"}

n = 100
model = xgb.train(
   params=params,
   dtrain=dtrain_reg,
   num_boost_round=n,
)

```

## Evaluate Model

```python
from sklearn.metrics import mean_squared_error

preds = model.predict(dtest_reg)
```

```python
rmse = mean_squared_error(y_test, preds, squared=False)

print(f"RMSE of the model: {rmse:.3f}")
```

```python
model.get_score(importance_type='gain')
```

# Example 2: SARIMA
How self-similar are median sales price trends?

```python
import matplotlib.pyplot as plt
import statsmodels.api as sm

region = 'San Antonio, TX'

df = input_df[(input_df['Region Type'] == 3) &
              (input_df['Region'] == region) &
              (input_df['Home Type'] == 2)][['Date', 'Median Sale Price']]

df['Log Median Sale Price'] = df['Median Sale Price'].apply(np.log)
df['Date'] = pd.to_datetime(df['Date'])
df = df.set_index(['Date'])
df = df.dropna()

plt.figure(figsize=(15,7))
plt.title("Log Median Sale Price in " + region)
plt.xlabel('Date')
plt.ylabel('Log Median Sale Price')
plt.plot(df['Log Median Sale Price'])
plt.show()
```

```python
df["Rolling Average"] = df["Log Median Sale Price"].rolling(window=52).mean()
df["Rolling Standard Deviation"] = df["Log Median Sale Price"].rolling(window=52).std()

plt.figure(figsize=(15,7))
plt.plot(df["Log Median Sale Price"], color='blue', label='Original')
plt.plot(df["Rolling Average"], color='red', label='Rolling Average')
plt.plot(df["Rolling Standard Deviation"], color='green', label='Rolling Standard Deviation')
plt.legend(loc='best')
plt.title('Rolling Mean and Standard Deviation')
plt.show(block=False)
```

```python
import pmdarima as pm
```

```python
df['time'] = [x for x in range(0, len(df))]
df = df.set_index('time')
df['Log Median Sale Price']

ARIMA_model = pm.auto_arima(
    df['Log Median Sale Price'],
    start_p=1,
    start_q=1,
    test='adf',
    max_p=15,
    max_q=15,
    # max_p=7,
    # max_q=7,
    # m=52,
    # d=None,
    d=1,
    # seasonal=True,
    trace=True,
    error_action='warn',
    # suppress_warnings=True,
    stepwise=True)
```

```python
ARIMA_model.plot_diagnostics(figsize=(15,12))
plt.show()
ARIMA_model
```

```python
def forecast(ARIMA_model, periods=104):
    fitted, confint = ARIMA_model.predict(n_periods=periods, return_conf_int=True)
    # index_of_fc = pd.date_range(df.index[-1] + pd.DateOffset(weeks=1), periods = periods, freq='W-SAT')
    index_of_fc = [x for x in range(df.index[-1], df.index[-1] + periods)]

    fitted_series = np.exp(pd.Series(fitted, index=index_of_fc))
    lower_series = np.exp(pd.Series(confint[:, 0], index=index_of_fc))
    upper_series = np.exp(pd.Series(confint[:, 1], index=index_of_fc))

    # Plot
    plt.figure(figsize=(15,7))
    plt.plot(np.exp(df["Log Median Sale Price"]), color='red')
    plt.plot(fitted_series, color='blue')
    plt.fill_between(lower_series.index,
                    lower_series,
                    upper_series,
                    color='blue', alpha=.15)

    plt.title("ARIMA/SARIMA - Forecast of Median Home Prices in " + region)
    plt.show()

forecast(ARIMA_model)
```
